__4. In a markdown block, outline the Machine Learning Linear Regression algorithm.__

Linear Regression involves determining a linear pattern for a collection of bivariate data (a set of data, each data points has two variables). The linear regression algorithm will calculate a line of best fit (linear equation/model) which can be used to predict new values based on either of the two variables.

For example, we can collect sample data on students amount of study hours and their corresponding assessment results. The Linear Regression algorithm can be used to determine a linear equation which models the pattern (association / relationship). __5. Modify the code block below to adjust $m$ and $b$ to manually fit a Linear Regression line with good intuition or fit.__

````python
x_train = np.array([[-4], [-8], [-10], [-12], [-14]])
y_train = np.array([1, 3, 5, 7, 9])

m = -0.8
b = -2.5

tmp_f_mb = compute_model_output(x_train, m, b,)

plt.plot(x_train, tmp_f_mb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Example Scatter Plot and Regression Line")
# Set the y-axis label
plt.ylabel('Dependent Variable')
# Set the x-axis label
plt.xlabel('Independent Variable')
plt.legend()

plt.show()
````

__7. Refering to the below model, describe underfitting in Linear Regression.__

````python
x_train = np.array([1, 1.5, 1.8, 2, 2.2])
y_train = np.array([[200], [250], [260], [280], [500]])


m = 200
b = 10

tmp_f_mb = compute_model_output(x_train, m, b,)

plt.plot(x_train, tmp_f_mb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.show()
````

### Underfitting
Underfitting is when the regression model doesn't sufficiently model the relationship.
This can occur for multiple reasons such as:
* Non-representative data
* Poor algorithm to determine the regression model
* The data doesn't actually have a pattern that can be well modelled
